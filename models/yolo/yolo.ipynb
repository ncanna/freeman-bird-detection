```python
class YOLODetector:
    def __init__(self,
                 model_name="yolo11n.pt",
                 experiment_name="yolo11_h03",
                 data_config="../../data/h03/h03.yaml",
                 data_dir="../../data/",
                 runs_dir = "../../outputs/runs"):
        self.model = YOLO(model_name)
        self.experiment_name = experiment_name
        self.data_config = Path(data_config).resolve()
        self.data_dir = Path(data_dir).resolve()
        self.runs_dir = Path(runs_dir).resolve()
        self.runs_dir.mkdir(parents=True, exist_ok=True)
        settings.update({"runs_dir": str(self.runs_dir),
                         "datasets_dir": str(self.data_dir),
                         "tensorboard": False,
                         "wandb": True})
        print(f"Model type: {self.model.model_name}")
        print(f"Using data config {self.data_config}")
        print(f"Runs will be saved to: {self.runs_dir}/{experiment_name}")
        self.experiment_name = experiment_name
        
    def train(self, epochs=1, imgsz=640, name=None):
        if name is None:
            name = f"{self.experiment_name}_train"
        self.model.train(data=self.data_config,
                         epochs=epochs,
                         imgsz=imgsz,
                         project=str(self.runs_dir),
                         name=name)

    def detect(self, test_data_path):
        results = self.model.predict(test_data_path)
        return results
```

```python
detector = YOLODetector()
```

```python
# Plot bounding boxes on each image
for result in results[0:10]:
    # Create annotator for the original image
    ann = Annotator(result.orig_img.copy())
    
    # Get boxes from the result
    boxes = result.boxes
    
    # Draw each bounding box
    for box in boxes:
        # Get box coordinates (xyxy format)
        xyxy = box.xyxy[0].cpu().numpy()
        
        # Get confidence score and class ID
        conf = box.conf[0].cpu().numpy()
        cls = int(box.cls[0].cpu().numpy())
        
        # Get class name from the model
        class_name = result.names[cls]
        
        # Create label with class name and confidence
        label = f"{class_name} {conf:.2f}"
        
        # Draw bounding box with label
        ann.box_label(xyxy, label)
    
    # Get the annotated image
    annotated_img = ann.result()
    
    # Convert BGR to RGB (OpenCV uses BGR, matplotlib expects RGB)
    annotated_img_rgb = cv2.cvtColor(annotated_img, cv2.COLOR_BGR2RGB)
    
    # Display the image
    plt.figure(figsize=(12, 8))
    plt.imshow(annotated_img_rgb)
    plt.axis('off')
    plt.tight_layout()
    plt.show()
```
