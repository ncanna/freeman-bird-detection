# Explore annotated video (H23)


```python
import json
from ultralytics.data import YOLODataset
```

# 1. Check the jason file is in COCO format

```python
path = '/home/hice1/wzhu97/scratch/2026_freeman_bird_behavior/data/annotated'
labels = f'{path}/H23.json'
video = ''

with open(labels) as f:
    data = json.load(f)

print(data.keys())
# Expected: dict_keys(['info', 'licenses', 'images', 'annotations', 'categories'])
# Minimum required: 'images', 'annotations', 'categories'

# All good!
```

# 2. Check content of each sections

```python
# Images: each must have id, file_name, width, height
print('\n# ##### images #####')
print('N=', len(data['images']), sep='')
print(data['images'][0])
# eg. {'id': 1, 'file_name': 'frame_001.jpg', 'width': 1920, 'height': 1080, ...}

# Annotations: each must have id, image_id, category_id, bbox, area, iscrowd
print('\n# ##### annotations #####')
print('N=', len(data['annotations']), sep='')
print(data['annotations'][0])
# {'id': 1, 'image_id': 1, 'category_id': 1,
#  'bbox': [x, y, width, height],  ← COCO uses xywh, NOT xyxy
#  'area': 500, 'iscrowd': 0}

# Categories: each must have id, name
print('\n# ##### categories #####')
print('N=', len(data['categories']), sep='')
print(data['categories'])
# [{'id': 1, 'name': 'fire', 'supercategory': 'none'}]
```
